### The goodreads_books_poetry.json file doesn't have title for each poetry, therefore we need to extract the title from the complete dataset goodread_books.json file, which is about 9GB.

```python
import pandas as pd
from pandas import Series, DataFrame
import json
```

```python
desciptionFile = pd.read_json('C:/Users/90312/Desktop/670/goodreads_books.json', lines=True,chunksize=1000)
```

```python
poetryFile = pd.read_json('C:/Users/90312/Desktop/670/goodreads_books_poetry.json', lines=True)
```

```python
bookId = set()
for index, row in poetryFile.iterrows():
    bookId.add(row['book_id'])
```

```python
dict = {}
#j = 0
for chunk in desciptionFile:
    for index, row in chunk.iterrows():
        if row['book_id'] in bookId:
            dict[str(row['book_id'])] = row['title']
            #print('yes')
    #j = j + 1
    #print(j)
```

```python
print(len(dict))
```

#### Store the extracted dictionary into a json file.

```python
# You need to use your own directory to store the poetryTitle.json file
with open('C:/Users/90312/Desktop/670/poetryTitle.json', 'w') as f:
    json.dump(dict,f)
```
