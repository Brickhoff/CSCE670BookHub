```python
#calculation the cosine similarity using sentence vector of description
string_pair = []
import numpy as np
import math
def takeSecond(elem):
        return elem[1]
    
def similarity(book_id):
    list_sim = [] #similarity with the target book
    index_target = list_bookid.index(book_id)
    
    norm_target = np.linalg.norm(data_infersent[index_target]['description'])
    if (norm_target == 0):
      cos = 0
    else:
      for e in range(len(list_bookid)):
        if(data_infersent[e]['description']==0):
          data_infersent[e]['description'] = [0]*4096
        cos = np.dot(data_infersent[e]['description'], data_infersent[index_target]['description']) / (norm_target * np.linalg.norm(data_infersent[e]['description']))
        list_sim.append(cos)    
        #print("e",e,"cos",cos)
    list_pair=[]
    pair= zip(list_bookid, list_sim)
    for e in pair:
      #if(e[1]<0.99):
      list_pair.append(e)    
    list_pair.sort(key= takeSecond, reverse=True)
    
    print('-------similarity rank--------')
    print('')
    print('    book_id   ','similarity(Sentence2Vec)' )
    string = [0]*len(list_pair)
        
    for index in range(len(list_pair)):
        string[index]= str(list_pair[index]).replace(")", ' ')
        string[index]= string[index].replace("(", ' ')
        string[index]= string[index].replace("'", '')
        string[index]= string[index].replace(",", ' ')
        print(index+1, string[index])
        string_pair.append(list_pair[index])
        if index == 15:
            break
similarity('25330489')
```

```python
#calculate similarity for each two books in the list. 

def similarity_revise (book_id):#given a book_id of target book, return 10 top books 
    list_sim = [] #similarity with the target book
    index = list_bookid.index(book_id)
    matrix = data_infersent[index]['description']# matrix of target book 
    
    divider_target=0
    for num in range(50):
        divider_target = pow(matrix[num],2) + divider_target
    divider_target = math.sqrt(divider_target)    
    
    for e in range(len(data_infersent)):#each book
        numerator = 0
        divider = 0
        cos = 0
        for i in range(50):
            numerator = numerator + matrix[i]*data_infersent[e]['description'][i]
            divider = divider + pow(data_infersent[e]['description'][i],2)
        divider = math.sqrt(divider) 
        if(divider != 0 and divider_target!=0):
            cos = numerator/(divider*divider_target)
        list_sim.append(cos)
    #print(list_sim[0])  
    #print(list_sim[1]) 
    
    #pair the similarity and the book, rank the similarity
    list_pair=[]
    pair= zip(list_bookid, list_sim)
    for e in pair:
        list_pair.append(e)    

    list_pair.sort(key= takeSecond, reverse=True)
    #print(list_pair[0])
    
    print('-------similarity rank--------')
    print('')
    print('    book_id   ','similarity(Word2Vec)' )
    list_similar = []
    for index in range(10):
        list_similar.append(list_pair[index])
    print(list_similar[0])        
    return list_similar
print(similarity_revise("1370171"))
```

```python
#If similarity between them is high then screen out the one which rank lower.
def filter(list_total):
    for index in range(16):
        list_similar = []
        list_similar = similarity_revise(str(list_total[index][0]))#返回一个rank 1 书的相似列表,长度为10
        print(index)
        for book in list_total[index+1:]:
            for similar_book in list_similar:
                if(str(book[0]) == similar_book[0]):
                    print("false")
                    list_total.remove(book)
    return list_total
print(len(filter(list_total)))
```
